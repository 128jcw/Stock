```python
df=pd.read_csv('카카오_뉴스타이틀.csv')
df
```

```python
import re
import numpy as np

okt = Okt()
noun_list = []
title_rename = []
for i in range(len(df)):
    title_rename.append(re.sub("[\(\[].*?[\)\]]", "",df.iloc[i]['기사제목']))
    noun_list.append(' '.join(okt.nouns(df.iloc[i]['기사제목'])))
df['nouns'] = noun_list
df['기사제목']=title_rename
df = df[df['nouns']!='']

#2차 불용어 제거
df['기사제목'] = df['기사제목'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['기사제목'].replace('', np.nan, inplace=True)
df = df.dropna(how='any')
df
```

```python
vocab={}
cnt=0
for i in df['nouns']:
    i=i.split(' ')
    for j in range(len(i)):
        if i[j] in vocab or len(i[j])<=1:
            cnt=cnt+1
            pass
        else:
            vocab[i[j]]=0
print(len(vocab), vocab)
vocab2=vocab
```

```python
up=6200
down=6200
up_rate=up/(up+down)
down_rate=down/(up+down)

import collections
for i,w in enumerate(df['nouns']):
    w = w.split(' ')
    if (df.iloc[i]['주가변동']==1):
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
                continue
            vocab[noun] = vocab[noun] + down_rate
    else:
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
                continue
            vocab[noun] = vocab[noun] - up_rate
print(vocab)            
f=open('카카오_단어_감성점수.txt', 'w')
for code, name in vocab.items():
    if code=='카카오':
        f.write(f'{code} : -3.0')
    else:
        f.write(f'{code} : {name}\n')
f.close()
```

```python
total=[]
for i, w in enumerate(df['nouns']):
    score=0
    w=w.split(' ')
    for j in w:
        if len(j)<=1:
            continue
        elif j not in vocab:
            continue
        elif j=='카카오':
            score=score-3
        else:
            score=score+vocab[j]
    total.append(score/len(w))
df['감성점수']=total
df
```

```python
df.to_csv('카카오_뉴스감성점수.csv', index=False, encoding='utf-8')
```
