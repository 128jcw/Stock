```python
df=pd.read_csv('카카오_뉴스타이틀.csv')
df
```

```python
import re
import numpy as np

okt = Okt()
noun_list = []
title_rename = []
for i in range(len(df)):
    title_rename.append(re.sub("[\(\[].*?[\)\]]", "",df.iloc[i]['기사제목']))
    noun_list.append(' '.join(okt.nouns(df.iloc[i]['기사제목'])))
df['nouns'] = noun_list
df['기사제목']=title_rename
df = df[df['nouns']!='']

#2차 불용어 제거
df['기사제목'] = df['기사제목'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['기사제목'].replace('', np.nan, inplace=True)
df = df.dropna(how='any')
df
```

```python
vocab={}
cnt=0
for i in df['nouns']:
    i=i.split(' ')
    for j in range(len(i)):
        if i[j] in vocab or len(i[j])<=1:
            cnt=cnt+1
            pass
        else:
            vocab[i[j]]=0
print(len(vocab), vocab)
vocab3=vocab
```

```python
import collections

for i,w in enumerate(df['nouns']):
    count=0
    w = w.split(' ')
    if (df.iloc[i]['주가변동']==1):
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
                continue
            vocab3[noun] = vocab3[noun] + 1
    else:
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
                continue
            vocab3[noun] = vocab3[noun]
print(len(vocab3), vocab3)
```

```python
for x in vocab3:
    count=0
    for i, w in enumerate(df['nouns']):
        w=w.split(' ')
        for j in w:
            if x==j:
                count=count+1
    vocab3[x]=vocab3[x]/count
print(vocab3)
```

```python
vocab4={}
for x in vocab3:
    if not(vocab3[x]==1.0 or vocab3[x]==0.0):
        vocab4[x]=vocab3[x]
print(len(vocab4), vocab4)
f=open('카카오_단어_감성점수3.txt', 'w')
for code, name in vocab4.items():
    f.write(f'{code} : {name}\n')
f.close()
```

```python
total=[]
for i, w in enumerate(df['nouns']):
    score=0
    count=0
    w=w.split(' ')
    for j in w:
        if len(j)<=1:
            continue
        elif j not in vocab4:
            continue
        else:
            score=score+vocab4[j]
            count+=1
    if count==0:
        total.append(4.5)
    else:
        total.append(score/count)
df['감성점수']=total
df
```

```python
df.to_csv('카카오_뉴스감성점수3.csv', index=False, encoding='utf-8')
```
